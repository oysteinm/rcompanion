Function to calculate the final value after reducing by a fixed percentage for z years

```r
reduce_by_percentage <- function(x, percentage, z) {
  if (z == 0) {
    return(x)  # When years = 0, return the initial value without reduction
  }
  for (i in 1:z) {
    x <- x * (1 - percentage / 100)
  }
  return(x)
}
```

Example usage:
Base value x = 100, percentage reduction = 20%, for z = 3 years

```r
final_value <- reduce_by_percentage(100, 20, 3)
print(final_value)
```

What is the total percentage reduction for 3 years?

```r
total_percentage_reduction <- 100*(1 - final_value / 100)
print(total_percentage_reduction)
```

Make a plot of starting value vs final value for different years, including year 0 = 100

```r
years <- 0:10
values <- sapply(years, function(y) reduce_by_percentage(100, 20, y))
```

Plot the results

```r
plot(years, values, type = "o", xlab = "Years", ylab = "Final Value", main = "Final Value vs Years", ylim = c(0, 100))
```
